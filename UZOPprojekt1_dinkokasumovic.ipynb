```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = pd.read_csv("diabetes.csv")
x
```

```python
x.info()
```

```python
x.describe()
```

```python
cols = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
x[cols].replace({'0':np.nan, 0:np.nan})
```

```python
x.describe()
```

```python
x.size
```

```python
x.shape
```

```python
x[x.Outcome == 1].shape
```

```python
x.nunique()
```

```python
x.isna().sum()
```

```python
x["Insulin"].fillna(x["Insulin"].median(), inplace=True)
x["SkinThickness"].fillna(x["SkinThickness"].median(), inplace=True)

x["BloodPressure"].fillna(x["BloodPressure"].median(), inplace=True)

x.dropna(subset = ["Glucose"], inplace=True)
x.dropna(subset = ["BMI"], inplace=True)
```

```python
x.isna().sum()
```

```python
corr = x.corr()
sns.heatmap(corr, annot=True, square=True)
plt.show()
```

```python
x.groupby("Outcome").mean()
```

```python
x.groupby("Outcome").mean().T.plot(figsize=(12,4))
```

```python
sns.pairplot(x,hue='Outcome',palette = ["#ff8811","#3339FF"])
plt.show()
```

```python
plt.hist(x.Age)
plt.xlabel("Age")
plt.show()
```

```python
plt.boxplot(x.Insulin)
plt.show()
```

```python
plt.scatter(x.Insulin, x.Glucose)
plt.show()
```
